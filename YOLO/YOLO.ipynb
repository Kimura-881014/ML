```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

# 物体検出

```python
# Load a model
model = YOLO('yolov8n.pt')

# Predict the model
img = model.predict('https://ultralytics.com/images/bus.jpg', save=False, conf=0.5)
```

```python
img = img[0].plot()
img  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```
