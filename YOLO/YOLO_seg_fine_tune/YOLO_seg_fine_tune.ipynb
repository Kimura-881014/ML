# YOLOv8 train

```python
!pip install ultralytics
```

```python
from ultralytics import YOLO

model = YOLO('yolov8l-seg.pt')
```

```python
model.train(
    data = './datasets/YOLODataset/dataset.yaml',
    epochs = 100,
    batch = 16,
    workers = 8,
    imgsz = 640,
    project = 'runs/pylon/train',
    name = 'v1',
    )
```

```python
from IPython.display import Image,display_png

display_png(Image('./runs/pylon/train/v18/results.png'))
```

```python
metrics = model.val()
```

## predict

```python
model = YOLO("./runs/pylon/train/v18/weights/best.pt")

results = model.predict(source="./datasets/YOLODataset/images/val/")
```

```python
from matplotlib import pyplot as plt
from PIL import Image
```

```python
for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
```
